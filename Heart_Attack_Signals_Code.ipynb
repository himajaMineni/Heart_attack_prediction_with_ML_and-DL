```python
#importing librosa for loading and analysing wave files
import librosa as lr


import IPython.display as ipd
```

```python
#loading a single wave file to analyze
file=("H:/research/classification of heart sound challenge/Thesis_dataset/Training set/training-a/a0001.wav")

#loading the file using librosa library, here the audio file is converted into a numpy array,
# and the sampling rate is by default 22KHz.
amp, sfre= lr.load(file)
```

```python
# to play the audio

#importing IPython.display to play the audio file
import IPython.display as ipd

#playing the audio file, we loaded using librosa
ipd.Audio(file)
```

```python
#importing display method from librosa library to display the plots 
import librosa.display
```

```python
#plotting the array of the audio file
plt.figure(figsize=(21,3))
lr.display.waveplot(amp,sfre)
```

```python
#Here we are finding the short time fourier transform of the audio clip
# It gives the amplitude of a different frquencies at a given timeperiod.
stft_of_file = lr.stft(amp)
stft_of_file
```

```python

#Converting amplitude of short time fourier transform signal to dB Scaled signal
dB_scaled = lr.amplitude_to_db(abs(stft_of_file))
dB_scaled
```

```python
#setting the figure size
plt.figure(figsize=(21, 3))

#displaying the spectrogram plot of the converted stft file in hertzs
lr.display.specshow(dB_scaled, sr=sfre, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
#set the figure size
plt.figure(figsize=(21,3))

#displaying the spectrogram plot of the converted stft file as a logrithmic function
lr.display.specshow(dB_scaled, sr=sfre, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
#importing the sklearn for modelling  
import sklearn
```

```python

#finding the Spectral centroid of the audio file
centroid_points=lr.feature.spectral_centroid(amp,sr=sfre)[0]
centroid_length = len(centroid_points)
centroid_length

```

```python

#set the figure size to 21*3
plt.figure(figsize=(21,3))

#finding the range of the centroid-points
frame_range = range(centroid_length)

frame_range
```

```python

# For visualizing calculating the time period of the sound wave
time_period = lr.frames_to_time(frame_range)

time_period
```

```python

#creating a scaling function for finding the min and max points of a signal
def scaling(amp, axis=0):
    normal = sklearn.preprocessing.minmax_scale(amp,axis=axis)
    return normal
```

```python
#set the figure size
plt.figure(figsize=(21,3))

#the Spectral Centroid plot of the wavefile
lr.display.waveplot(amp, sr=sfre, alpha=0.5)
plt.plot(time_period, scaling(centroid_points))

```

```python

#finding the spectral rolloff of the wavefile
rolloff_points = lr.feature.spectral_rolloff(amp+0.07,sr=sfre)[0]

#set the figure size
plt.figure(figsize=(21,3))

#displaying the waveplot of spectral_rolloff
lr.display.waveplot(amp,sr=sfre, alpha=0.8)

#plot of time_period vs rolloff_points
plt.plot(time_period, scaling(rolloff_points))
```

```python

# finding the bandwidths of the wavefile
BW1= lr.feature.spectral_bandwidth(amp+0.08, sr=sfre)[0]

BW2= lr.feature.spectral_bandwidth(amp+0.08, sr=sfre, p=2)[0]

```

```python

#set the figure size
plt.figure(figsize=(21,3))

#displaying the bandwidth plot 
lr.display.waveplot(amp, sr=sfre, alpha= 0.7)


#plot of bandwidth vs time_period
plt.plot(time_period, scaling(BW1))

plt.plot(time_period, scaling(BW2))

plt.legend(('p=1','p=2'))
```

```python

# zoom in parameters
x1 = 6000
x2 = 9000

#set the figure size of the plot
plt.figure(figsize=(21,3))

#ploting the wavefile with Zoom in parameters
plt.plot(amp[x1:x2])

#plot the grid
plt.grid()
```

```python

#calculating zero crossing rate 
zero_crossing_rate = lr.zero_crossings(amp[x1:x2], pad=False)
count_of_zero_crossings = sum(zero_crossing_rate)
count_of_zero_crossings

```

```python
length_of_hop = 1900
```

```python

#calcualting the chroma vector of the wavefile
chroma_vector = lr.feature.chroma_stft(amp, sr=sfre, hop_length=length_of_hop) 
chroma_vector
```

```python

#Finding the Mel frequency ceptral coefficients of the wavefile to study the entire shape of the spectrum
mel_freq_cc = lr.feature.mfcc(amp, sr= sfre)
mel_freq_cc
```

```python

#displaying the spectrogram plot of the mel frequency ceptral coefficients
plt.figure(figsize=(21,3))
lr.display.specshow(mel_freq_cc, sr=sfre, x_axis= 'time')
```

```python
from sklearn.decomposition import PCA


training_pca = PCA(n_components = 2)
training_pca.fit(X)

transform_pca = training_pca.transform(X)

```

```python
variation_of_each_component = training_pca.explained_variance_ratio_
total_variation = sum(training_pca.explained_variance_ratio_)

print("Variance of each pca_components: ",variation_of_each_component)
print("Total variance of pca_components: ",total_variation)

plt.figure(figsize = (8,6))
plt.scatter(transform_pca[:,0], transform_pca[:,1], c=y, cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
